### 3. Write a Python program to count the number 4 in a given list.

List = [1,4,6,8,4,9,4]

```python
def list_count_4(nums):
    count = 0  
    for num in nums:
        if num == 4:
            count = count + 1
            return count
        print(list_count_4([1,4,6,8,4,9,4]))
```

### 9. Write a Python function that takes a list and returns a new list with unique elements of the first list.

Sample List : [1,2,3,3,3,3,4,5]

Unique List : [1, 2, 3, 4, 5]

```python
def unique_list(l):
    x = []
    for a in l:
        if a not in x:
            x.append(a)
            return x
        print(unique_list([1,2,3,3,3,3,4,5])) 
```

### 1. Write a Python program to concatenate all elements in a list into a string and return it.

```python
def concatenate_list_data(list):
    result= ''
    for element in list:
        result += str(element)
    return result

print(concatenate_list_data(['Hii','This','is','Deepali']))
```

```python
# A)change all the categorical columns into numerical by creating Dummies and using label encoder.

bank[['job','marital','education','default','housing','loan','contact','poutcome','month',
       'y']] = bank[['job','marital','education','default','housing','loan','contact','poutcome','month', 'y']].astype('category')
bank1 = bank

bank1.info()
```

```python
#Label Encoding

bank1['month'] = bank1['month'].cat.codes
bank1['job'] = bank1['job'].cat.codes
bank1['marital'] = bank1['marital'].cat.codes
bank1['education'] = bank1['education'].cat.codes
bank1['default'] = bank1['default'].cat.codes
bank1['housing'] = bank1['housing'].cat.codes
bank1['loan'] = bank1['loan'].cat.codes
bank1['contact'] = bank1['contact'].cat.codes
bank1['poutcome'] = bank1['poutcome'].cat.codes
bank1['y'] = bank1['y'].cat.codes
```

```python
pd.get_dummies(bank1)
```

```python
# B) rename all the column names DF

bank1.columns = [['AGE','JOBS','MARITAL','EDUCATION','DEFAULT','BALANCE','HOUSING','LOAN','CONTACT','DAY','MONTH','DURATION',
               'CAMPAIGN','PDAYS','PREVIOUS','POUTCOME','Y']] 

bank1.head(4)
```

```python
# C) Rename only one specific column in DF

bank1 = bank1.rename(columns={'AGE': 'age'})
bank1.head(3)
```

# Web Scrapping

### 5. Write a python program to scrape reviews from a commercial web site

- Solution: Scrapping Amazon website reviews for product 'OnePlus-Mirror-Black-128GB-Storage' using 'BeautifulSoup'

```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
link = 'https://www.amazon.in/OnePlus-Mirror-Black-128GB-Storage/product-reviews/B07DJHV6VZ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
```

```python
page = requests.get(link)
page
```

```python
page.content
```

```python
soup = bs(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
names = soup.find_all('span',class_='a-profile-name')
names
```

```python
cust_name = []
for i in range(0,len(names)):
    cust_name.append(names[i].get_text())
cust_name
```

```python
cust_name.pop(0)
```

```python
cust_name
```

```python
title = soup.find_all('a',class_='review-title-content')
title
```

```python
review_title = []
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
review_title
```

```python
review_title[:] = [titles.lstrip('\n') for titles in review_title]
review_title
```

```python
review_title[:] = [titles.rstrip('\n') for titles in review_title]
review_title
```

```python
rating = soup.find_all('i',class_='review-rating')
rating
```

```python
rate = []
for i in range(0,len(rating)):
    rate.append(rating[i].get_text())
rate
```

```python
len(rate)
```

```python
rate.pop(0)
```

```python
rate
```

```python
review = soup.find_all("span",{"data-hook":"review-body"})
review
```

```python
review_content = []
for i in range(0,len(review)):
    review_content.append(review[i].get_text())
review_content
```

```python
review_content[:] = [reviews.lstrip('\n') for reviews in review_content]
review_content
```

```python
review_content[:] = [reviews.rstrip('\n') for reviews in review_content]
review_content
```

```python
len(review_content)
```

```python
df = pd.DataFrame()
```

```python
df['Customer Name']=cust_name
df['Review title']=review_title
df['Ratings']=rate
df['Reviews']=review_content
```

```python
df
```

```python
df.to_csv(r'Amazon_reviews_scrapping.csv',index=True)
```
